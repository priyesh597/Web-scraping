```python
html = urlopen(url)
```

```python
all_links = soup.find_all('a')
for link in all_links:
    print(link.get('href'))
```

```python
rows = soup.find_all('tr')
a = rows[:10]
for i in a:
    print(i)
```

```python
for row in a:
    row_td = row.find_all('td')
print(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells,'lxml').get_text()
cleantext
```

```python
import re 
list_rows =[]
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean,'',str_cells))
    print(clean2)
```
